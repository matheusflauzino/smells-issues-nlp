```python
import csv
import os
import pandas as pd
import numpy as np
```

```python
PATH_REPOSITORIES = 'repositories'
PATH_LOC = 'loc_results'
PATH_RESULTS = 'results'
FILE_LIST_REPOSITORIES = 'data/filtered_data.csv'
FILE_LOC = 'data/loc_repositories_total.csv'
FILE_DOWN = 'data/rep_down.csv'
FILE_SMELLS = 'data/smells.csv'
FILE_ISSUES = 'data/issues.csv'
```

```python
downloads = pd.read_csv(FILE_DOWN)
```

```python
locs = pd.read_csv(FILE_LOC)
```

```python
smells = pd.read_csv(FILE_SMELLS)
```

```python
issues = pd.read_csv(FILE_ISSUES)
```

```python
issues
```

```python
df = smells.loc[:,'name':'smells']
df['loc'] = pd.Series(locs['loc'],index=df.index)
```

```python
df['commits']  = pd.Series(downloads['commits'],index=df.index)
df['link']  = pd.Series(downloads['link'],index=df.index)
```

```python
df['issues'] = pd.Series(issues['open'],index=df.index)
df['bugs'] = pd.Series(issues['bugs'],index=df.index)
df['no_bugs'] = pd.Series(issues['no_bugs'],index=df.index)
```

```python
dataset = pd.DataFrame(df,index=None,columns=['name','link','loc','dc','lc','lm','lpl','tmm','smells','issues','bugs','no_bugs'])
```
